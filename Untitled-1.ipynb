```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

from scipy.interpolate import interp1d
T1 = 2*np.pi
lambda_func = lambda t: 5*np.sin(t)
n = 1000 #liczba punkt√≥w do szacowania funkcji odwrotnej

def niejednorodny_p_pois_odwrotna_dys(lambda_func, T, n):
    m_T = scipy.integrate.quad(lambda_func, 0, T)[0]
    N_t = np.random.poisson(m_T)
    ts = np.linspace(0,T,n)
    m_values = [scipy.integrate.quad(lambda_func, 0, t)[0] for t in ts]
    m_inv = interp1d(m_values, ts)
    us = np.random.rand(N_t)
    Ts1 = m_inv(m_T * us)
    Ts1.sort()
    return Ts1


def rysowanie_procesu_poissona(T, skoki, lambda_func):
    x_i = [0]
    y_i = [0]
    xs = np.linspace(0, T, 1000)
    for i in range(len(skoki)):
        x_i.append(skoki[i])
        y_i.append(y_i[-1] + 1)

        plt.vlines(x=x_i[-1], ymin=y_i[-1] - 1, ymax=y_i[-1], linestyles='dashed', linewidth=4)
        plt.hlines(y=y_i[-2], xmin=x_i[-2], xmax=x_i[-1], linewidth=4)
        plt.scatter(x=[x_i[-1]], y=[y_i[-2]], facecolors='none', edgecolors='b', s=80)
        plt.scatter(x=[x_i[-1]], y=[y_i[-1]], c='b', s=65)
        plt.plot(xs, lambda_func(xs))

    plt.hlines(y=y_i[-1], xmin=x_i[-1], xmax=T, linewidth=4)

    plt.show()
    
Ts = niejednorodny_p_pois_odwrotna_dys(lambda_func, T1,n)
rysowanie_procesu_poissona(T1, Ts, lambda_func)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def emp_characteristic_fun(sample, t):
    sample = np.array(sample)
    t = np.array(t)
    echf = np.mean(np.exp((1j) * t[:, None] * sample), axis=1)
    return echf

#rozk≈Çad jednostajny U(-1,1)
U = np.random.uniform(-1,1, size = 10**3)
t = np.linspace(-20,20, 10**4)
#teoretycznie
tchf = np.sin(t)/t
tchf = (np.exp(-1j*t) - np.exp(1j*t))/(-2j*t)
#estymator
echf = emp_characteristic_fun(U, t)

plt.plot(t, tchf.real, label = "teoretyczna")
plt.plot(t, echf.real, label = "estymator")
plt.title('rozk≈Çad jednostajny U(-1,1)')
plt.legend()

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry rozk≈Çadu chi2
k = 4

# Generowanie pr√≥bki z rozk≈Çadu gamma
samples = np.random.chisquare(k,10000)

# Obliczanie warto≈õci funkcji charakterystycznej dla rozk≈Çadu normalnego
t_values = np.linspace(0, 10, 100)
chf_chi2 = ((1-2*1j*t_values)**(-k/2))

# Obliczanie pr√≥bkowego estymatora funkcji charakterystycznej
F = np.zeros_like(t_values)
G = np.zeros_like(t_values)
i = -1
for t in t_values:
    i += 1
    F[i] = np.mean(np.exp(1j * t * samples))
    G[i] = np.imag(np.mean(np.exp(1j * t * samples)))

plt.subplot(2,1,1)
plt.title("Czesc rzeczywista")
plt.legend(["teorytyczny", "pr√≥bkowy estymator"])
plt.plot(t_values, chf_chi2, t_values, F)
plt.subplot(2,1,2)
plt.title("Czesc urojona")
plt.plot(t_values, np.imag(chf_chi2), t_values, G)
plt.legend(["teorytyczny", "pr√≥bkowy estymator"])
plt.tight_layout()
plt.show()
```

```python
def Poisson_normal(T,lam):
    k = np.random.poisson(T*lam)
    U = np.random.uniform(0,T,k)
    U = sorted(U)
    wielkosci = np.random.standard_normal(len(U))
    Y = np.cumsum(wielkosci)

    plt.step(U, Y, where = 'pre')

Poisson_normal(30,2)
```
