```python
import scipy.linalg as alg
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from mpl_toolkits.mplot3d import Axes3D 
```

```python

def Marsylia(N):
    X = np.random.uniform(-1,1,N)
    Y = np.random.uniform(-1,1,N)
    P = []
    for i in range(N):
        s = X[i]**2+Y[i]**2
        if s <= 1:
            P.append((X[i]*np.sqrt(((-2)*np.log(s))/s), (Y[i]*np.sqrt(((-2)*np.log(s))/s))))
    return P


P = Marsylia(100000)
X = []
Y = []
for i in range(len(P)):
    X.append(P[i][0])
    Y.append(P[i][1])

plt.hist2d(X,Y, bins = 100, range = [[-3,3],[-3,3]])




```

```python
def Marsylia3D(N):
    X = np.random.uniform(-1,1,N)
    Y = np.random.uniform(-1,1,N)
    Z = np.random.uniform(-1,1,N)
    P = []
    for i in range(N):
        s = X[i]**2+Y[i]**2+Z[i]**2
        if s <= 1:
            P.append((X[i]*np.sqrt(((-2)*np.log(s))/s), (Y[i]*np.sqrt(((-2)*np.log(s))/s)), (Z[i]*np.sqrt(((-2)*np.log(s))/s))))
    return P

P = Marsylia3D(1000)
X = [P[i][0] for i in range(len(P))]
Y = [P[i][1] for i in range(len(P))]
Z = [P[i][2] for i in range(len(P))]


Cov = [[50,32,-4],[32,34,0],[-4,0,80]]
A = alg.cholesky(Cov)
X1 = [A @ P[i] for i in range(len(P))]

fig = plt.figure(figsize=(10, 10)) 
ax = Axes3D(fig) 

for i in range(len(X1)):
    img = ax.scatter(X1[i][0], X1[i][1], X1[i][2], marker='s', 
                 s=100, color='gray') 

plt.show() 
```
